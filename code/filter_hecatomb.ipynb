```python
import pandas as pd
import numpy as np
import seaborn as sns
from collections import Counter
import matplotlib.pyplot as plt
```

### Viruses in long-term migration

```python
data = pd.read_csv('../data/hecatomb/long_term_migration/bigtable.tsv.gz', sep = '\t', compression = 'gzip')
samples = list(set(data['sampleID']))
```

```python
prefilter_df = pd.DataFrame() 
filter_df = pd.DataFrame() 
virus_df = pd.DataFrame() 
level = 'species'

for s in samples: 
    
    this_sample = data[data['sampleID'] == s]
    this_sample['prefilter_percent'] = this_sample['count']*100/np.sum(this_sample['count'])
    this_sample_viruses = this_sample[this_sample['kingdom'] == 'Viruses']
    this_sample_level = pd.DataFrame(this_sample_viruses.groupby(level)['prefilter_percent'].sum())
    this_sample_level.columns = [s]

    
    # filter - filters for amino acid hits and least common ancestor method
    this_sample_filter = this_sample[this_sample['alnType'] == 'aa']
    this_sample_filter = this_sample_filter[this_sample_filter['taxMethod'] == 'LCA']
    this_sample_filter['filter_percent'] = this_sample_filter['count']*100/np.sum(this_sample_filter['count'])
    this_sample_filter_viruses = this_sample_filter[this_sample_filter['kingdom'] == 'Viruses']
    this_sample_filter_level = pd.DataFrame(this_sample_filter_viruses.groupby(level)['filter_percent'].sum())
    this_sample_filter_level.columns = [s]

    # filter for viruses only 
    this_sample_viruses = this_sample_filter[this_sample_filter['kingdom'] == 'Viruses'] 
    this_sample_viruses['virus_percent'] = this_sample_viruses['count']*100/np.sum(this_sample_viruses['count'])
    this_sample_viruses_level = pd.DataFrame(this_sample_viruses.groupby(level)['virus_percent'].sum())
    this_sample_viruses_level.columns = [s]
    
    #append dataframes 
    prefilter_df = pd.concat([prefilter_df,this_sample_level.T])
    filter_df = pd.concat([filter_df,this_sample_filter_level.T])
    virus_df = pd.concat([virus_df,this_sample_viruses_level.T])
```

```python
virus_df.T.to_csv('../output_files/long_term_migration_virus_hecatomb_species.tsv', sep = '\t')
```
